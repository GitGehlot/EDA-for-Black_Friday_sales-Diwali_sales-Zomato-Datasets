```python
df_country=pd.read_excel('/content/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_final=pd.merge(df,df_country,on='Country Code',how='left')
df_final.head(2)
```

```python
df_final.columns
```

```python
country_names=df_final['Country'].value_counts().index
```

```python
country_val=df_final['Country'].value_counts().values
```

```python
#pie chart for top 3 countries based on transactions

plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Zomato maximum records is from India, then USA and follows by UK

```python
df_final.columns
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text','Country']).size().reset_index().rename(columns={0:'Rating Columns'})
```

```python
ratings
```

Observation
1. When Rating is between 4.5 to 4.9 > Excellent
2. When Rating are between 4.0 to 3.4-->very good
3. when Rating is between 3.5 to 3.9 --> good
4. when Rating is between 3.0 to 3.4 -> average
5. when Rating is between 2.5 to 2.9--> average
6. when Rating is between 2.0 to 2.4--> Poor

```python
ratings.head()
```

```python
#sns.set(rc={'figure.figsize':(12,6)})  #or via matplotlib
plt.rcParams['figure.figsize']=(12,6)
a1=sns.barplot(x='Aggregate rating',y='Rating Columns',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
#count graph

sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])

```
